## Algebric Manipulation
#### i. Simplification

```python
from sympy import simplify, Symbol
x=Symbol("x")
expr=(x**2+2*x+1)/(x+1)
simplified=simplify(expr)
print(simplified)
```

#### ii. Expansion and factorization

```python
from sympy import expand, factor
expand((x+1)**2)
factor(x**2+2*x+1)
```

## Calculus 
#### i. Differentiation

```python
from sympy import diff
diff(x**2+2*x+3*x**2)
```

#### ii. Integration

```python
from sympy import integrate 
integrate(x**2+2*x+4*x+4**x)
```

#### iii. Limits

```python
from sympy import limit 
limit((x**6+2*x+3)/(2*x+1),x,1)
```
