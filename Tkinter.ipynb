# Creating interactive application

```python
import tkinter as tk

def update_label():
    text=entry.get()
    label.config(text=f"Hello,{text}")

root=tk.Tk()
root.title("My application")

entry=tk.Entry(root)
entry.grid(row=1,column=3)

button=tk.Button(root,text="Update label",command=update_label)  
button.grid(row=2,column=3)

label=tk.Label(root,text="Enter your name")
label.grid(row=3,column=3)

root.mainloop()
```

# Basic Calculator 

```python
import tkinter as tk

def calculate():
    try:
        result=eval(entry.get())
        entry.delete(0,tk.END)
        entry.insert(tk.END,str(result))
    except:
        entry.delete(0,tk.END)
        entry.insert(tk.END,"Error")

def clear():
    entry.delete(0,tk.END)

def button_click(symbol):
    entry.insert(tk.END,symbol)

# creat tkinter window
window=tk.Tk()
window.title("Basic calculator")

#Entry widget to display input & output
entry=tk.Entry(window,width=30,font=("Calibri",16))
entry.grid(row=0,column=0,columnspan=4,padx=10,pady=10)

# define button
buttons=[
    ('7',1,0),  ('8',1,1),   ('9',1,2),  ('/',1,3),
    ('4',2,0),  ('5',2,1),   ('6',2,2),  ('*',2,3),
    ('1',3,0),  ('2',3,1),   ('3',3,2),  ('-',3,3),
    ('0',4,0),  ('C',4,1),   ('=',4,2),  ('+',1,3)
]

# Create button and assign event handlers
for (symbol,row,column) in buttons:
    if symbol == 'C':
        btn=tk.Button(window,text=symbol,width=5,height=2,command=clear)
    elif symbol == '=':
        btn=tk.Button(window,text=symbol,width=10,height=4,command=calculate)
    else:
        btn=tk.Button(window,text=symbol,width=5,height=2,command=lambda s=symbol:button_click(s))
    btn.grid(row=row,column=column,padx=5,pady=5)

window.mainloop()
               
```
