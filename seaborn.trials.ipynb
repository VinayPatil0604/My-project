```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df
```

```python
df.columns
```

```python
sns.histplot(x='sepal_length',hue='species',data=df,kde=True)
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=df)
```

```python
sns.boxplot(y='sepal_length',x='species',data=df)
```

```python

sns.heatmap(df.drop('species',axis=1).corr(),annot=True)
```

```python
sns.set_style('white') 
sns.pairplot(df,hue='species')
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',hue='species',data=df)
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=df)
```

```python
sns.stripplot(x='species',y='petal_length',data=df)
```

```python
sns.violinplot(x='sepal_length',hue='species',data=df)
```

```python
sns.swarmplot(x='species',y='petal_length',data=df)
```

```python
sns.countplot(x='species',data=df)
```
